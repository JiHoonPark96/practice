```python
import pandas as pd

# Load the datasets
bs_df = pd.read_stata("C:\\Users\\starw\\OneDrive\\바탕 화면\\Compustat (MS)\\BS(NA).dta")
cf_df = pd.read_stata("C:\\Users\\starw\\OneDrive\\바탕 화면\\Compustat (MS)\\CF(NA).dta")
global_df = pd.read_stata("C:\\Users\\starw\\OneDrive\\바탕 화면\\Compustat (MS)\\global.dta")
is_df = pd.read_stata("C:\\Users\\starw\\OneDrive\\바탕 화면\\Compustat (MS)\\IS(NA).dta")
```

```python
# Create a summary of all variables in the dataframe
variable_summary = pd.DataFrame({
    'Variable': is_df.columns,
    'DataType': is_df.dtypes,
    'Missing (%)': is_df.isnull().mean() * 100,
    'Unique': is_df.nunique(),
    'Mean': is_df.mean(numeric_only=True),
    'Std': is_df.std(numeric_only=True)
})

# Sort by missing percentage
variable_summary = variable_summary.sort_values('Missing (%)')

# Display the first 20 most complete variables
print("Most Complete Variables (First 20):")
print(variable_summary.head(20))

# Display basic statistics for key financial variables
key_vars = ['ni', 'sale', 'revt', 'ebit', 'ebitda', 'xint', 'pi']
print("\nKey Financial Variables Summary:")
print(is_df[key_vars].describe())
```

```python
# Create a summary of all variables in the cf_df dataframe
cf_variable_summary = pd.DataFrame({
    'Variable': cf_df.columns,
    'DataType': cf_df.dtypes,
    'Missing (%)': cf_df.isnull().mean() * 100,
    'Unique': cf_df.nunique(),
    'Mean': cf_df.mean(numeric_only=True),
    'Std': cf_df.std(numeric_only=True)
})

# Sort by missing percentage
cf_variable_summary = cf_variable_summary.sort_values('Missing (%)')

# Display the first 20 most complete variables
print("Most Complete Variables in Cash Flow Dataset (First 20):")
print(cf_variable_summary.head(20))

# Display basic statistics for key cash flow variables
key_cf_vars = ['oancf', 'ivncf', 'fincf', 'capx', 'dv', 'sstk', 'prstkc']
print("\nKey Cash Flow Variables Summary:")
print(cf_df[key_cf_vars].describe())
```

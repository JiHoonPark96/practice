```python
# Cell 2: imports and device setup

import os
import json

import numpy as np
import pandas as pd

import torch
from torch.utils.data import Dataset

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, f1_score, mean_squared_error
from sklearn.linear_model import LinearRegression

from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    Trainer,
    TrainingArguments,
)

from peft import LoraConfig, get_peft_model

print("PyTorch version:", torch.__version__)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
# Cell 3: Load THOR / FrameBERT JSON files

frame_path = r"C:\Users\starw\Downloads\data_2012_frame_thor_output.json"
business_path = r"C:\Users\starw\Downloads\data_business_thor_output_1200.json"

df_frame = pd.read_json(frame_path, lines=False)
df_business = pd.read_json(business_path, lines=False)

print("df_frame shape:", df_frame.shape)
print("df_business shape:", df_business.shape)

print("\n=== df_frame columns ===")
print(df_frame.columns)

print("\n=== df_business columns ===")
print(df_business.columns)

print("\n=== df_frame head ===")
display(df_frame.head())

print("\n=== df_business head ===")
display(df_business.head())
```

```python
# Cell 4: Basic cleaning and label distribution

# Ensure required columns exist
required_cols = ["headline", "polarity"]
for col in required_cols:
    if col not in df_business.columns:
        raise ValueError(f"Column '{col}' not found in df_business")

# Drop rows with missing headline or polarity
df_business_clean = df_business.dropna(subset=["headline", "polarity"]).copy()

print("Original business rows:", len(df_business))
print("Clean business rows:", len(df_business_clean))

print("\nPolarity value counts (string):")
print(df_business_clean["polarity"].value_counts())
```

```python
# Cell 5: Build training DataFrame from THOR-labeled business data

# Define mapping from polarity string to integer label
polarity2id = {
    "negative": 0,
    "neutral": 1,
    "positive": 2
}

# Filter to only rows with polarity in mapping
df_train_thor = df_business_clean[df_business_clean["polarity"].isin(polarity2id.keys())].copy()

df_train_thor["text"] = df_train_thor["headline"].astype(str)
df_train_thor["label"] = df_train_thor["polarity"].map(polarity2id)

print("THOR training dataset size:", len(df_train_thor))
print("\nLabel distribution (int):")
print(df_train_thor["label"].value_counts())

display(df_train_thor[["text", "polarity", "label"]].head())
```

```python
# Cell 6: Optional - Load FinancialPhraseBank and combine

use_phrasebank = True  # False로 두면 THOR만 사용

df_phrasebank = None

if use_phrasebank:
    try:
        from datasets import load_dataset

        ds = load_dataset("takala/financial_phrasebank", "sentences_allagree")
        # ds is a DatasetDict with a "train" split
        ds_train = ds["train"]

        # Convert to DataFrame
        df_phrasebank = pd.DataFrame(ds_train)
        # Expected columns: 'sentence', 'label' (string: 'positive', 'negative', 'neutral')

        # Some versions have 'sentence'/'label' or 'sentence'/'category', adjust if needed
        text_col = "sentence"
        label_col = "label"
        if text_col not in df_phrasebank.columns or label_col not in df_phrasebank.columns:
            print("Unexpected columns in PhraseBank, got:", df_phrasebank.columns)
        else:
            df_phrasebank["text"] = df_phrasebank[text_col].astype(str)
            df_phrasebank["polarity"] = df_phrasebank[label_col].astype(str)
            df_phrasebank = df_phrasebank[df_phrasebank["polarity"].isin(polarity2id.keys())].copy()
            df_phrasebank["label"] = df_phrasebank["polarity"].map(polarity2id)

            print("PhraseBank size:", len(df_phrasebank))
            print(df_phrasebank["label"].value_counts())
    except Exception as e:
        print("Could not load PhraseBank dataset. Reason:", e)
        use_phrasebank = False
        df_phrasebank = None
else:
    print("Skipping PhraseBank loading.")
```

```python
# Cell 7: Prepare final training DataFrame and train/val split

if use_phrasebank and df_phrasebank is not None:
    df_data = pd.concat(
        [df_train_thor[["text", "label"]], df_phrasebank[["text", "label"]]],
        ignore_index=True
    )
else:
    df_data = df_train_thor[["text", "label"]].copy()

print("Final training dataset size:", len(df_data))
print(df_data["label"].value_counts())

train_df, val_df = train_test_split(
    df_data,
    test_size=0.1,
    random_state=42,
    stratify=df_data["label"]
)

print("\nTrain size:", len(train_df), "Val size:", len(val_df))
print("Train label dist:\n", train_df["label"].value_counts())
print("Val label dist:\n", val_df["label"].value_counts())
```

```python
# Cell 8: Tokenizer and custom Dataset

model_name = "ProsusAI/finbert"

tokenizer = AutoTokenizer.from_pretrained(model_name)

class FinancialSentimentDataset(Dataset):
    def __init__(self, df, tokenizer, max_length=128):
        self.texts = df["text"].tolist()
        self.labels = df["label"].tolist()
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = int(self.labels[idx])

        enc = self.tokenizer(
            text,
            truncation=True,
            padding="max_length",
            max_length=self.max_length,
            return_tensors="pt",
        )
        item = {k: v.squeeze(0) for k, v in enc.items()}
        item["labels"] = torch.tensor(label, dtype=torch.long)
        return item

train_dataset = FinancialSentimentDataset(train_df, tokenizer)
val_dataset = FinancialSentimentDataset(val_df, tokenizer)

print("Train dataset length:", len(train_dataset))
print("Val dataset length:", len(val_dataset))
```

```python
# Cell 9: Load FinBERT and apply LoRA

base_model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=3
)

# Typical LoRA config for BERT-like models
lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    lora_dropout=0.1,
    target_modules=["query", "value"],  # applies to attention modules
    task_type="SEQ_CLS",
)

model = get_peft_model(base_model, lora_config)
model.to(device)

model.print_trainable_parameters()
```

```python
# Cell 10: Trainer setup

output_dir = "./finbert_lora_thor"

training_args = TrainingArguments(
    output_dir=output_dir,
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    learning_rate=2e-5,
    eval_strategy="epoch",  # Changed from evaluation_strategy
    save_strategy="epoch",
    load_best_model_at_end=True,
    logging_steps=50,
    logging_dir=os.path.join(output_dir, "logs"),
)

def compute_metrics(pred):
    logits = pred.predictions
    labels = pred.label_ids
    preds = logits.argmax(-1)

    acc = accuracy_score(labels, preds)
    f1 = f1_score(labels, preds, average="macro")

    return {"accuracy": acc, "macro_f1": f1}

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
)

trainer
```

```python
# Cell 11: Train and evaluate

train_result = trainer.train()
print(train_result)

eval_metrics = trainer.evaluate()
print("Evaluation metrics:", eval_metrics)
```

```python
# Cell 12: Detailed classification report on validation set

pred_output = trainer.predict(val_dataset)
logits = pred_output.predictions
y_true = pred_output.label_ids
y_pred = logits.argmax(-1)

print("Accuracy:", accuracy_score(y_true, y_pred))
print("Macro F1:", f1_score(y_true, y_pred, average="macro"))
print("\nClassification report:")
print(classification_report(
    y_true,
    y_pred,
    target_names=["negative", "neutral", "positive"]
))
```

```python
# Cell 13: Save fine-tuned LoRA model and tokenizer

save_dir = "./finbert_lora_thor_saved"
os.makedirs(save_dir, exist_ok=True)

trainer.model.save_pretrained(save_dir)
tokenizer.save_pretrained(save_dir)

print("Model and tokenizer saved to:", save_dir)
```

```python
# Cell 14: Helper function for sentiment prediction

id2polarity = {v: k for k, v in polarity2id.items()}

def predict_sentiment(texts, max_length=128):
    if isinstance(texts, str):
        texts = [texts]

    enc = tokenizer(
        texts,
        truncation=True,
        padding=True,
        max_length=max_length,
        return_tensors="pt"
    ).to(device)

    model.eval()
    with torch.no_grad():
        logits = model(**enc).logits

    preds = logits.argmax(-1).cpu().numpy().tolist()
    labels = [id2polarity[p] for p in preds]
    return labels

# Quick test
example_texts = [
    "The company's profits soared after the merger.",
    "The market crashed due to unexpected inflation data.",
    "The board announced no major changes in the near term."
]
print(predict_sentiment(example_texts))
```

```python
# Cell 19 (revised): Load HuffPost financial news (auto-detect fields)

try:
    import kagglehub
except ImportError:
    import sys
    !{sys.executable} -m pip install kagglehub -q
    import kagglehub

import os
import json
import pandas as pd

# 1) KaggleHub에서 데이터 다운로드
path = kagglehub.dataset_download("irakozekelly/financial-news-sentiment-dataset-20122022")
print("Downloaded folder path:", path)

# 2) JSON 파일 수집
json_files = sorted([f for f in os.listdir(path) if f.endswith(".json")])
print("JSON files detected:", json_files)

all_records = []

# 3) JSON 로드
for fname in json_files:
    fpath = os.path.join(path, fname)
    with open(fpath, "r", encoding="utf-8") as f:
        data = json.load(f)

        # 구조가 {"root": [...] } 일 수도 있고 list일 수도 있음
        if isinstance(data, dict) and "root" in data:
            items = data["root"]
        elif isinstance(data, list):
            items = data
        else:
            print("⚠ Unknown JSON structure:", fname)
            continue

        all_records.extend(items)

print("Total raw items:", len(all_records))

# 4) DataFrame 생성
news_df = pd.DataFrame(all_records)
print("Columns available:", list(news_df.columns))

# 5) 날짜 컬럼 자동 탐지
possible_date_cols = ["date", "date_published", "published", "publish_date", "time", "created_at"]

date_col = None
for col in possible_date_cols:
    if col in news_df.columns:
        date_col = col
        break

if date_col is None:
    raise ValueError("❌ No usable date column found in dataset. Available columns: " + str(list(news_df.columns)))

print("Detected date column:", date_col)

# 6) 날짜 파싱
news_df["date"] = pd.to_datetime(news_df[date_col], errors="coerce").dt.date
news_df = news_df.dropna(subset=["date"])

# 7) 헤드라인 컬럼 자동 탐지
possible_headline_cols = ["headline", "title", "news_title", "short_description"]

headline_col = None
for col in possible_headline_cols:
    if col in news_df.columns:
        headline_col = col
        break

if headline_col is None:
    raise ValueError("❌ No usable headline column found.")

print("Detected headline column:", headline_col)

news_df["headline"] = news_df[headline_col].astype(str)
news_df = news_df[["date", "headline"]].copy()

print("Final usable rows:", len(news_df))
display(news_df.head())
```

```python
# Cell 20: Batch-safe FinBERT-LoRA sentiment prediction

import torch
import numpy as np

def predict_sentiment_batch(text_list):
    """
    안정적으로 batch 감성예측 수행:
    - NaN/None 제거
    - padding=True, truncation=True 적용
    - 리스트/단일 문자열 모두 처리
    """

    # 단일 문자열 처리
    if isinstance(text_list, str):
        text_list = [text_list]

    # 모든 입력을 문자열로 정제
    clean_list = []
    for x in text_list:
        if x is None:
            continue
        try:
            clean_list.append(str(x))
        except:
            continue

    if len(clean_list) == 0:
        return []

    # tokenizer batch encode
    inputs = tokenizer(
        clean_list,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=128
    ).to(device)

    # 모델 추론
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        pred_ids = torch.argmax(logits, dim=1).cpu().numpy()

    id2label = {0: "negative", 1: "neutral", 2: "positive"}
    preds = [id2label[p] for p in pred_ids]
    return preds



# ---------- 전체 뉴스에 대해 감성예측 ----------
headlines = news_df["headline"].astype(str).tolist()
n_total = len(headlines)
print("Total headlines to predict:", n_total)

batch_size = 64
all_preds = []

for i in range(0, n_total, batch_size):
    batch = headlines[i : i + batch_size]
    preds = predict_sentiment_batch(batch)
    all_preds.extend(preds)

    if (i // batch_size) % 50 == 0:
        print(f"Processed {min(i + batch_size, n_total)} / {n_total}")

# 결과 저장
news_df["sent_label"] = all_preds
label2score = {"negative": -1, "neutral": 0, "positive": 1}
news_df["sent_score"] = news_df["sent_label"].map(label2score)

print("\nSentiment label distribution:")
print(news_df["sent_label"].value_counts())
display(news_df.head())
```

```python
# Cell 21: Create daily sentiment index

# 날짜 범위 price_df 기준으로 제한
min_date = price_df["date"].min()
max_date = price_df["date"].max()
news_df = news_df[(news_df["date"] >= min_date) & (news_df["date"] <= max_date)].copy()

print("Usable news rows in date range:", len(news_df))

# 일별 감성 집계
sent_daily = (
    news_df.groupby("date")
    .agg(
        sent_index=("sent_score", "mean"),
        n_headlines=("sent_score", "count")
    )
    .reset_index()
)

print("Rows before filtering:", len(sent_daily))
display(sent_daily.head())

# 하루 뉴스 개수 최소 threshold
min_headlines = 5
sent_daily = sent_daily[sent_daily["n_headlines"] >= min_headlines].copy()

print(f"\nRows after filtering (>= {min_headlines} headlines/day):", len(sent_daily))
print("\nSentiment index stats:")
print(sent_daily["sent_index"].describe())
display(sent_daily.head())
```

```python
# Cell 22: Merge sentiment with market returns and build predictive model

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

if sent_daily.empty:
    print("ERROR: sent_daily is empty! Try reducing min_headlines or check news_df.")
else:
    # 가격 데이터와 병합
    merged_df = pd.merge(
        sent_daily,
        price_df[["date", "ret"]],
        on="date",
        how="inner"
    ).sort_values("date").reset_index(drop=True)

    print("Rows after merging with price data:", len(merged_df))
    display(merged_df.head())

    # 5일 누적 수익률
    horizon = 5
    merged_df["ret_next_5d"] = merged_df["ret"].shift(-1).rolling(window=horizon).sum()
    merged_df = merged_df.dropna(subset=["ret_next_5d"]).reset_index(drop=True)

    print("\nRows after building forward return:", len(merged_df))
    display(merged_df.head())

    # 회귀 (sent_index → 5일 다음 수익률)
    X = merged_df[["sent_index"]].values
    y = merged_df["ret_next_5d"].values

    reg = LinearRegression()
    reg.fit(X, y)
    y_hat = reg.predict(X)

    r2 = reg.score(X, y)
    rmse = np.sqrt(mean_squared_error(y, y_hat))
    corr = np.corrcoef(y, y_hat)[0, 1]

    print("\nRegression results (5-day forward):")
    print("R^2:", r2)
    print("RMSE:", rmse)
    print("Correlation:", corr)
```
